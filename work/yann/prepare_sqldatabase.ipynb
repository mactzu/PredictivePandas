```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import numpy as np
```

```python
beers=pd.read_csv("../../data/beer.gz")
beers.head()
```

```python
beers=pd.DataFrame(beers.groupby(["brewery_id","brewery_name","beer_beerid","beer_name","beer_style"]).agg(
    {"beer_abv":"mean",
     "review_overall":"mean",
     "review_aroma":"mean",
     "review_appearance":"mean",
     "review_palate":"mean",
     "review_taste":"mean"}))
beers=beers.reset_index(drop=False)
```

```python
words=pd.read_csv("reviews.csv")
words=words.drop(["brewery_id","review_overall"],axis=1)
beers=beers.merge(words,on="beer_beerid")
beers.columns
```

```python
brewery=beers[["brewery_id","brewery_name"]]
brewery=brewery.drop_duplicates(subset=["brewery_id"],keep='first')
brewery.columns=["id","name"]
brewery.head()
#brewery.to_csv("../database/brewery.csv",index=False)
```

```python
beer_style=pd.DataFrame({"style":beers["beer_style"].unique()})
beer_style.head()
#beer_style.to_csv("../database/beer_style.csv",index=False)
```
